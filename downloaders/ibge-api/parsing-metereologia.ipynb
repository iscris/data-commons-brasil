```python
import pandas as pd
```

```python
import json

# Open the file
outputs_list = []
with open('dump-metereologia-mensal.jsonl', 'r') as file:
    # Iterate over each line in the file
    for line in file:
        # Parse the JSON object from the line
        json_object = json.loads(line.strip())
        # Process the JSON object (for example, print it)
        df = pd.DataFrame.from_dict(json_object)

        
        for index,row in df.iterrows():
            if (len(row[0]["01"]) > 0):
                for month in ['01', '02', '03', '04', '05', '06', '07', '08', '09', '10', '11', '12']:
                    for data_point in row[0][month]:
                        # print()
                        data = {
                                'dtHrLeitura': data_point['dtHrLeitura'],
                                'pressaoAtm': data_point['pressaoAtm'],
                                'temperaturaExt': data_point['temperaturaExt'],
                                'umidadeExt': data_point['umidadeExt'],
                                'direcaoVento': data_point['direcaoVento'],
                                'velocidadeVento': data_point['velocidadeVento'],
                                'precipitacao': data_point['precipitacao'], 
                                'place': df.columns[0]
                            }
                        outputs_list.append(data)
                        
output_parsed = pd.DataFrame(outputs_list)
```

```python
output_parsed[output_parsed["place"]=="EMIMB"]
# output_parsed
```

```python
# substituindo pela informação do estado onde fica o maregrafo
maregrafos_dict = {
    "EMBEL": "PA",
    "EMSAN": "AP",
    "EMFOR": "CE",
    "EMSAL": "BA",
    "EMARC": "RJ",
    "EMIMB": "SC"
}

output_parsed['place'] = output_parsed['place'].replace(maregrafos_dict)
```

```python
# agora salvar um csv por variavel
for coluna in ['pressaoAtm', 'temperaturaExt', 'umidadeExt', 'direcaoVento', 'velocidadeVento', 'precipitacao']:
    subset_df = output_parsed[['dtHrLeitura','place',coluna]]
    subset_df.to_csv("metereologia-"+coluna+".csv",index=False)
    print(subset_df.head())

# output_parsed.to_csv("metereologia.csv", index=False)
```
